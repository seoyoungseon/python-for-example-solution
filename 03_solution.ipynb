````
#작업 디렉토리에 python이란 이름의 파일이나 디렉토리가 있는지 확인하고 있으면 삭제하시오.
dir
del python
#이름이 python인 하위 디렉토리를 만드시오.
mkdir python
#python 디렉토리에 이름이 src, data, backup인 세 개의 하위 디렉토리를 만드시오.
cd python
mkdir src data backup
#작업 디렉토리를 src 디렉토리로 이동하시오.
cd src
#메모장으로 작업 디렉토리에 hello_python.py 파일을 작성하시오. 이 파일에 다음 한 줄을 입력하고 저장하시오.
notepad hello_python.py
#hello_python.py를 복사하여 backup 디렉토리에 hello_python_backup.py란 이름으로 저장하시오.
copy hello_python.py ..\backup\hello_python_backup.py
#작업 디렉토리에 어떤 파일이 있는지 확인하시오.
dir
#작업디렉토리에 있는 hello_python.py를 삭제하시오.
del hello_python.py
#backup 디렉토리에 있는 hello_python_backup.py를 복사하여 작업 디렉토리에 hello_python.py란 이름으로 저장하시오.
copy ..\backup\hello_python_backup.py ..\src\hello_python.py
#hello_python.py를 usb 메모리의 최상위 디렉토리로 이동하시오.
move hello_python.py J:\
#작업 디렉토리를 usb 메모리의 최상위 디렉토리로 이동하시오.
cd ..\..
#src, data, backup 디렉토리를 삭제하시오.
rmdir .\python\src
rmdir .\python\data
rmdir .\python\backup
#hello_python.py를 python 디렉토리로 이동하시오.
move hello_python.py .\python
#작업 디렉토리를 python 디렉토리로 이동하시오.
cd python
#다음 명령을 실행하시오.
ipython hello_python.py
````
