1. **대사율자료**에대한 아래 분석 요구를 처리하되 가능하면 `comprehension`을 사용하시오.

```
gender = ['M']*2 + ['F']*4 + ['M'] + ['F']*2 + ['M'] + ['F']*2 + ['M'] + ['F']*4 + ['M']*2
weight = [62.0, 62.9, 36.1, 54.6, 48.5, 42.0, 47.4, 
          50.6, 42.0, 48.7, 40.3, 33.1, 51.9, 42.4,
          34.5, 51.1, 41.2, 51.9, 46.9]
metabolic_rate = [1792, 1666, 995, 1425, 1396, 1418, 1362, 1502,
        1256, 1614, 1189, 913, 1460, 1124, 1052, 1347,
        1204, 1867, 1439]
```

    * 남자(여자)의 체중과 대사율을 추출하여 별도의 리스트에 저장하시오.

```
m_weight = [ weight[idx] for idx, g in enumerate(gender)
            if g == 'M' ]
m_metabolic_rate = [ metabolic_rate[idx] for idx, g in enumerate(gender)
                    if g == "M"]

f_weight = [ weight[idx] for idx, g in enumerate(gender)
              if g == 'F']
f_metabolic_rate = [ metabolic_rate[idx] for idx, g in enumerate(gender)
                      if g == 'F']
print(m_weight)
print(m_metabolic_rate)
print(f_weight)
print(f_metabolic_rate)
```

    * 체중과 대사율의 합, 평균, 편차, 표준편차를 성별로 구하시오.

```
import numpy as np
```

```
#남자 체중
sum_m_wgt = sum(m_weight)
mean_m_wgt = np.mean(m_weight)
std_m_wgt = np.std(m_weight)
sd_m_wgt = mean_m_wgt - m_weight #평균(값) - 데이터(list) -> 브로드캐스팅 : 편차(list)
#남자 대사율
sum_m_mr = sum(m_metabolic_rate)
mean_m_mr = np.mean(m_metabolic_rate)
std_m_mr = np.std(m_metabolic_rate)
sd_m_mr = mean_m_mr - m_metabolic_rate
#여자 체중
sum_f_wgt = sum(f_weight)
mean_f_wgt = np.mean(f_weight)
std_f_wgt = np.std(f_weight)
sd_f_wgt =  mean_f_wgt - f_weight
#여자 대사율
sum_f_mr = sum(f_metabolic_rate)
mean_f_mr = np.sum(f_metabolic_rate)
std_f_mr = np.std(f_metabolic_rate)
sd_f_mr = mean_f_mr - f_metabolic_rate

print(sum_m_wgt, mean_m_wgt, std_m_wgt, sd_m_wgt )
print(sum_m_mr, mean_m_mr, std_m_mr, sd_m_mr)
print(sum_f_wgt, mean_f_wgt, std_f_wgt, sd_f_wgt )
print(sum_f_mr, mean_f_mr, std_f_mr, sd_f_mr)
```

2. **심장병자료**에 대한 아래 분석 요구를 처리하되 가능하면 `comprehension`을 사용하시오.

```
country = ["Australia","Austria"," Belgium/Luxembourg","Canada"," Denmark",\
         "Finland","France","Iceland","Ireland","Italy",\
         "Netherlands","New Zealand","Norway","Spain","Sweden",\
         "Switzerland","United Kingdom","United States","West Germany"]
wine = [2.5, 3.9, 2.9, 2.4, 2.9, 0.8, 9.1, 0.8, 0.7, 7.9, 1.8, 1.9, 0.8, 6.5, 1.6, 5.8, 1.3, 1.2, 2.7]
death_rate = [211, 167, 131, 191, 220, 297, 71, 211, 300, 107, 167, 266, 227, 86, 207, 115, 285, 199, 172]
```

    * 연간 포도주 소비량이 5 이상인 국가의 이름과 사망률을 별도의 리스트에 저장하시오.

```
country_upper_5 = [ country[idx] for idx, w in enumerate(wine)
                    if w >= 5]
death_rate_upper_5 = [ death_rate[idx] for idx, w in enumerate(wine)
                      if w >= 5]
print(country_upper_5)
print(death_rate_upper_5)
```

    * 연간 포도주 소비량이 5 미만인 국가의 이름과 사망률을 별도의 리스트에 저장하시오.

```
country_under_5 = [ country[idx] for idx, w in enumerate(wine)
                    if w < 5]
death_rate_under_5 = [ death_rate[idx] for idx, w in enumerate(wine)
                      if w < 5]
print(country_under_5)
print(death_rate_under_5)
```

    * 연간 포도주 소비량이 5 이상인 국가의 사망률의 합, 평균, 편차, 표준편차를 구하시오.

```
sum_death_rate_ge_5 = sum(death_rate_upper_5)
mean_death_rate_ge_5 = np.mean(death_rate_upper_5)
sd_death_rate_ge_5 = mean_death_rate_ge_5 - death_rate_upper_5
std_death_rate_ge_5 = np.std(death_rate_upper_5)

print(sum_death_rate_ge_5, mean_death_rate_ge_5, sd_death_rate_ge_5, std_death_rate_ge_5)
```
