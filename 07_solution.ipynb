```
cm = 123456

inch = cm/2.54
inc=1
ft_in=inc*12
yd_ft=ft_in*3
mi_yd=yd_ft*1760

mi=inch//mi_yd
inch=inch%mi_yd

yd=inch//yd_ft
inch=inch%yd_ft

ft= inch//ft_in
inch=inch%ft_in

print(mi, "mile", yd, "yard", ft, "feet", inch, "inch")
```

8. $n=20$, $p=0.3$, $x=8$일 때 $f(x) = \binom{n}{x} p^x (1-p)^{n-x}$의 값을 계산하여 출력하시오.

```
import math
n = 20
p = 0.3
x = 8
fx = math.factorial(n)/(math.factorial(x)*math.factorial(n-x))
print(fx)
```

9.   $\alpha = 1.2$, $\beta=2.5$, $x=0$일 때 함수 $p(x) = \frac{1}{1 + e^{\alpha + \beta x}}$의 값을 계산하여 출력하시오.

```
import math
alpha = 1.2
beta = 2.5
x = 0
px = 1/(1 + math.exp(alpha + beta * x))
print(px)
```

12. `math` 모듈에 정의된 `asin()`, `acos()`, `atan()`, `degrees()`, `radians()`, `factorial()`, `isfinite()`, `isinf()`, `isnan()` 함수들에 대해 알아보고 이들을 사용하는 예제를 작성하시오.

```
x = math.pi
y = math.radians(x) # 라디안으로 변환
z = math.degrees(y)
print(z)

print(math.asin(y))
print(math.acos(y))

print(math.factorial(4)) #4!

print(math.isfinite(x)) #유한인지 아닌지 결과는 논리값
print(math.isinf(x)) #무한인지 아닌지 결과는 논리값
print(math.isnan(x)) #결측 인지 아닌지 결과는논리값
```
