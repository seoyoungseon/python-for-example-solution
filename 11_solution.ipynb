* 라디안을 도로 변환하는 프로그램을 작성하시오.

```
import math
radian=float(input("라디안: "))
degree = 180*radian/math.pi
print(degree)
```

* 평행사변형의 면적을 계산하는 프로그램을 작성하시오.

```
height = float(input("평행사변형의 높이: "))
base = float(input("평행사변형의 밑변: "))
area = height*base
print("면적: ", area)
```

* 원의 지름과 원호의 각도를 입력받아서 원호의 길이를 계산하는 프로그램을 작성하시오.

```
diameter = float(input("원의 지름: "))
circular_arcs_degree = float(input("원호의 각도: "))
circular_arcs_length = diameter/2*circular_arcs_degree
print("원호의 길이: ",circular_arcs_length)
```

* 입력받은 정수의 모든 약수를 찾아 출력하는 프로그램을 작성하시오.

```
num = int(input("정수: "))

for divisor in range(1,num+1):
  if (num%divisor==0):
    print(divisor)
```

* 0과 100 사이에서 난수 한 개를 발생시키는 프로그램을 작성하시오.

```
import random
print(random.randrange(0,100), end=" ")
```
