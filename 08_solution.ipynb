2.  체질량 지수(Body Mass Index, BMI)는 인간의 비만도를 나타내는 지수로, 체중과 키의 관계로 계산된다.
키가 t 미터, 체중이 w 킬로그램일 때, BMI는 $w / t^2$와 같이 계산한다.

    * 자신의 키와 체중이 저장된 변수를 정의하시오.
    * BMI를 계산해서 변수에 저장하시오.

```
t = 1.85
w = 70
bmi = w/t**2
```

3. 한국의 경우 BMI의 크기에 따라 다음과 같이 비만도를 분류한다.
BMI가 35 이상이면 고도 비만, 30 - 35 사이에 있으면 중등도 비만, 25 - 30 사이에 있으면 경도 비만, 23 - 24.9 사이에 있으면 과체중, 18.5 - 22.9 사이이면 정상, 18.5 미만이면 저체중으로 판정하고 있다.

    * 앞 문제의 키와 체중을 가진 사람은 어디에 해당하는지 판정해보시오.


```
if bmi >= 35:
  msg="고도 비만"
elif bmi>=30:
  msg="중등도 비만"
elif bmi>=25:
  msg="경도 비만"
elif bmi>=23:
  msg-"과체중"
elif bmi>=18.5:
  msg="정상"
else:
  msg="저체중"
print(msg)
```

4. 혈압은 심장 수축기의 최고 혈압과 이완기의 최저 혈압의 두 측정치로 요약되는데, 휴식시 정상 혈압은 수축시 100~140mmHg에 이완시 60~90mmHg이고, 혈압이 지속적으로 140/90mmHg 이상일 때 고혈압이 있다고 말한다.

    * 자신의 수축시 혈압(systolic blood pressure)과 이완시 혈압(diastolic blood pressure)이 저장된 변수를 정의하시오.
    * 정상 혈압인지 고혈압인지를 검사하는 프로그램을 작성하시오.

```
systolic_blood_pressure = 112
diastolic_blood_pressure = 65

if (systolic_blood_pressure >= 110) and (systolic_blood_pressure <= 140) and (diastolic_blood_pressure >= 60) and (diastolic_blood_pressure <= 90):
  msg = "정상혈압 입니다."
elif (systolic_blood_pressure > 140) and (diastolic_blood_pressure > 90):
  msg = "고혈압 입니다."
print(msg)
```

5. $x=-5$일 때 함수 $f(x) = \left\{
     \begin{array}{ll}
       x & \text{if } x \gt 0 \\
       0 & \text{if} x \le 0
     \end{array}
   \right.$를 계산하시오.
이런 함수는 rectified linear function이라고 한다.

```
x = -5

if (x > 0):
  fx = x
else:
  fx = 0
print(fx)
```

6.  $\alpha=2$, $x=-5$일 때 함수 $f(x) = \left\{
     \begin{array}{ll}
       x & \text{if } x \gt 0 \\
       \alpha (e^x - 1) & \text{if} x \le 0
     \end{array}
   \right.$를 계산하시오.
이런 함수는 exponential linear function이라고 한다.

```
import math
alpha = 2
x = -5

if x > 0:
  fx = x
else:
  fx = alpha*(math.exp(x)-1)
print(fx)
```

7. 모듈 `random`에 정의된 `randint()` 함수는 두 정수 사이에 있는 정수를 하나 무작위로 추출해주는 난수발생함수이다.
예를 들어 `randint(100, 999)`는 세 자리 정수 중에서 하나를 랜덤하게 뽑아준다.
모듈 `random`의 `randint()` 함수를 이용하여 두 자리 정수 하나를 무작위로 발생시킨 다음 이 수의 십 자리 수를 알아보는 프로그램을 작성하시오.
두 자리 정수의 십 자리 수를 알아보는 한 가지 방법은 10으로 나눌 때의 몫을 알아보는 것이다.

```
import random
x = random.randint(10,99)
print(x)
x_quotient = x//10

if x_quotient > 0:
  msg = x_quotient
print(msg)
```

8. 모듈 `random`의 `randint()` 함수를 이용하여 세 자리 정수 두 개를 랜덤하게 발생시킨 다음 그 중에서 큰 값을 찾아내는 프로그램을 작성하시오.

9. 모듈 random의 randint() 함수를 이용하여 세 자리 정수 세 개를 랜덤하게 발생시킨 다음 그 중에서 제일 큰 값을 찾아내는 프로그램을 작성하시오.

```
pop1 = random.randint(100,999)
pop2 = random.randint(100,999)
pop3 = random.randint(100,999)
print(pop1, pop2, pop3)

if pop1 > pop2:
  if pop1 > pop3:
    print(pop1)
elif pop2 > pop3:
  print(pop2)
else:
  print(pop3)
```
