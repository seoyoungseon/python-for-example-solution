3. 100 이하의 자연수 중에서 2의 배수 또는 3의 배수의 집합을 만들어 보시오.

````
num = list(range(1,101))

even = []
odd = []
for n in num:
  if n%2 == 0:
    even = even + [n]
  if n%3 == 0:
    odd = odd + [n]
   
even = set(even)
odd = set(odd)
print(even)
print(odd)
````
