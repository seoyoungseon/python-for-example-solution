* 1500 이상 2700 이하인 정수 중에서 5와 7의 배수를 찾는 프로그램을 작성하시오.

```
num = range(1500,2701)

n=[]
for i in num:
  if i%5==0 and i%7==0:
    n=n+[i]
print(n[:5])
```

* 여러 개의 정수가 저장된 리스트에서 짝수와 홀수의 개수를 세는 프로그램을 작성하시오.

```
num = list(range(1,10))
print(num)

even=[]
odd=[]

for i in num:
  if i%2==0:
    even=even+[i]
  else:
    odd=odd+[i]
print("짝수의 개수: ", len(even))
print("홀수의 개수: ", len(odd))
```

* 1부터 20까지의 정수를 출력하되 3의 배수는 수 대신 '삼', 5의 배수는 '오', 3과 5의 배수는 '삼오'를 출력하는 프로그램을 작성하시오.

```
for x in range(1,21):
  if x%3==0:
    print("삼")
    if x%5==0:
      print("삼오")
  elif x%5==0:
    print("오")
  else:
    print(x)  
```

* 영문자를 입력받아서 자음인지 모음인지 판단하는 프로그램을 작성하시오.

```
alphabet = input("영문자를 입력하세요: ")

vowel=["a","e","i"]
if alphabet.lower() in vowel:
  print("모음 입니다.")
else:
  print("자음 입니다.")
```
