1. $\lambda = 2$일 때 $0 \le x \le 20$ 구간을 1000 등분한 점에서 함수 $f(x) = 1 - e^{-\lambda x}$의 값을 계산하여 출력하시오.

```
import math
seq = list(range(1001))

x = []
for number in seq:
  x= x+[number/20]
print(x[:5])
  
lamda = 2.0
fx = []
for x0 in x:
  f = 1-(math.exp(-lamda*x0))
  fx = fx + [f]
print(fx[:5])
```

2. $n=20$, $p=0.3$일 때 $x=0, 1, \dots, n$에서 함수 $f(x) = \binom{n}{x} p^x (1-p)^{n-x}$의 값을 계산하여 출력하시오.

```
n = 20
p = 0.3
seq = list(range(0,n+1))

x = []
for number in seq:
  x = x + [number]
print(x)
  
fx = []
for x0 in x:
  f = (math.factorial(n)/(math.factorial(x0)*math.factorial(n-x0)))*(p**x0)*(1-(p**(n-x0)))
  fx = fx + [f]
print(fx[:5])
```

3. $\alpha = 1.2$, $\beta=2.5$일 때 $-10 \le x \le 10$ 구간을 2000 등분한 점에서 함수 $p(x) = \frac{1}{1 + e^{-\alpha - \beta x}}$의 값을 계산하여 출력하시오.

```
alpha = 1.2
beta = 2.5
seq = list(range(-10,11))

x = []
for number in seq:
  x= x+[number/2000]
print(x)

px = []
for x0 in x:
  f = 1/(1-math.exp(-alpha-beta*x0))
  px = px + [f]
  
print(px[:5])
```

4. 앞 문제에서 계산한 $p(x)$를 이용하여 $logit(x) = \log \left( \frac{p(x)}{1-p(x)} \right)$를 계산하여 출력하시오.

```
logit = []
for x in px:
  l = math.log(abs( x/(1-x) ))
  logit = logit + [l]
  
print(logit[:5])
```

5. 구간 $[-5, 5]$을 100 등분하는 점에서 함수 $f(x) = \left\{
     \begin{array}{ll}
       x & \text{if } x \gt 0 \\
       0 & \text{if} x \le 0
     \end{array}
   \right.$를 계산하시오.

```
seq = list(range(-5,6))

x = []
for number in seq:
  x = x + [number/100]
print(x)

for i in x:
  if i > 0:
    print(i)
  else:
    print(0)
```

6. $\alpha=2$일 때 구간 $[-5, 5]$을 100 등분하는 점에서 함수 $f(x) = \left\{
     \begin{array}{ll}
       x & \text{if } x \gt 0 \\
       \alpha (e^x - 1) & \text{if} x \le 0
     \end{array}
   \right.$를 계산하시오.

```
import math
seq = list(range(-5,6))

x = []
for number in seq:
  x = x + [number/100]
print(x)

alpha=2
for i in x:
  if i > 0:
    print(i)
  else:
    fx=alpha*(math.exp(i)-1)
    print(fx)
```

8. 한 사이트에 등록된 사용자는 5명이다.
5명 사용자의 이름과 암호가 저장된 리스트 usernames, passwords를 만드시오.

    * 사용자의 이름과 암호를 키보드에서 입력받아 정상적인 사용자인지를 확인하는 프로그램을 작성하시오.
    * 새로운 사용자의 이름과 암호를 입력받아서 usernames, passwords에 추가하는 프로그램을 작성하시오.

```
usernames = ["user1","user2","user3","user4","user5"]
passwords = ["123456","789465","144812","789456","151204"]

user_name = input("이름을 입력하시오: ")
password = input("암호를 입력하시오: ")

if user_name in usernames:
  if password in passwords:
    print("정상적인 사용자 입니다.")
  else:
    print("정상적인 사용자가 아닙니다.")
else:
    print("정상적인 사용자가 아닙니다.")
```
