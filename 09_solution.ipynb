1. 체질량 지수(Body Mass Index, BMI)는 인간의 비만도를 나타내는 지수로, 체중과 키의 관계로 계산된다.
키가 t 미터, 체중이 w 킬로그램일 때, BMI는 $w / t^2$와 같이 계산한다.

    * 키보드에서 키와 체중이 입력받아 적절하게 변환한 다음 변수에 저장하시오.
    * BMI를 계산해서 변수에 저장하시오.

```
msg_t = "키(m): "
msg_w = "체중(kg): "
msg_t = float(input(msg_t))
msg_w = float(input(msg_w))
bmi = msg_w/msg_t**2
print(bmi)
```

2. 한국의 경우 BMI의 크기에 따라 다음과 같이 비만도를 분류한다.
BMI가 35 이상이면 고도 비만, 30 - 35 사이에 있으면 중등도 비만, 25 - 30 사이에 있으면 경도 비만, 23 - 24.9 사이에 있으면 과체중, 18.5 - 22.9 사이이면 정상, 18.5 미만이면 저체중으로 판정하고 있다.

    * 키보드에서 입력받은 키와 체중을 가진 사람은 어디에 해당하는지 판정하는 프로그램을 작성하시오.

```
if bmi >= 35:
  msg="고도 비만"
elif bmi>=30:
  msg="중등도 비만"
elif bmi>=25:
  msg="경도 비만"
elif bmi>=23:
  msg-"과체중"
elif bmi>=18.5:
  msg="정상"
else:
  msg="저체중"
print(msg)
```

3. 혈압은 심장 수축기의 최고 혈압과 이완기의 최저 혈압의 두 측정치로 요약되는데, 휴식시 정상 혈압은 수축시 100~140mmHg에 이완시 60~90mmHg이고, 혈압이 지속적으로 140/90mmHg 이상일 때 고혈압이 있다고 말한다.

    * 키보드에서 수축시 혈압(systolic blood pressure)과 이완시 혈압(diastolic blood pressure)을 입력받은 다음 이를 적절히 변환해서 변수에 저장하시오.
    * 입력받은 혈압을 가진 사람이 정상인지 고혈압인지를 판정하는 프로그램을 작성하시오.

```
systolic_blood_pressure = "수축시 혈압(mmHg): "
diastolic_blood_pressure = "이완시 혈압(mmHg): "
systolic_blood_pressure = int(input(systolic_blood_pressure))
diastolic_blood_pressure = int(input(diastolic_blood_pressure))

if (systolic_blood_pressure >= 110) and (systolic_blood_pressure <= 140) and (diastolic_blood_pressure >= 60) and (diastolic_blood_pressure <= 90):
  msg = "정상혈압 입니다."
elif (systolic_blood_pressure > 140) and (diastolic_blood_pressure > 90):
  msg = "고혈압 입니다."
print(msg)
```

5.  키보드에서 $\alpha$와 $x$를 입력받아 함수 $f(x) = \left\{
     \begin{array}{ll}
       x & \text{if } x \gt 0 \\
       \alpha (e^x - 1) & \text{if} x \le 0
     \end{array}
   \right.$를 계산하시오.

```
import math
alpha = "숫자를 입력하시오.: "
x = "숫자를 입력하시오.: "
alpha = float(input(alpha))
x = float(input(x))

if x>0:
  fx = x
else:
  fx = alpha*(math.exp(x)-1)
print(fx)
```
